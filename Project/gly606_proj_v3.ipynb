```python
df_res['Tot'] = df_res['WAIS'] + df_res['EAIS'] + df_res['Peninsula']
df_res['Tot_slr'] = -df_res['Tot'] * (917/1000) * (1/(.71*np.pi*(6370e3)**2)) * 1000
#df_res_exp5 = df_res['Exp'] == 'exp05'

df_res_exp05 = df_res.loc[df_res['Exp'] == 'exp05']

labels_exp05 = df_res_exp05['Center+Model']
WAIS_exp05 = df_res_exp05['WAIS']
EAIS_exp05 = df_res_exp05['EAIS']
Pen_exp05 = df_res_exp05['Peninsula']
tot_exp05 = df_res_exp05['Tot']
tot_slr_exp05 = df_res_exp05['Tot_slr']

exp05_dict = {
    'WAIS': WAIS_exp05,
    'EAIS': EAIS_exp05,
    'Peninsula': Pen_exp05}

x = np.arange(len(labels_exp05))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots()

for attribute, measurement in exp05_dict.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    #ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('SLR (mm)')
ax.set_title('ISMIP6 AIS EXP05')
#ax.set_xticks(x + width, labels_exp05)
#ax.legend(loc='upper left', ncols=3)
#ax.set_ylim(0, 250)

plt.show()
```

```python
# Plot total SLR for each participant exp05 (STD melt) and exp13 (PIGL melt)

df_res_exp13 = df_res.loc[df_res['Exp'] == 'exp13']
tot_slr_exp13 = df_res_exp13['Tot_slr']
labels_exp13 = df_res_exp13['Center+Model']


fig, ax = plt.subplots()

ax.bar(labels_exp05, tot_slr_exp05, label=labels_exp05, align='edge', width=0.5)

ax.set_ylabel('mm SLR')
ax.set_title('ISMIP6 AIS EXP05 (Std melt) 2100 SLR')
#ax.legend(title='Model+Center')

plt.xticks(rotation = 45, rotation_mode='anchor', ha='right')
plt.ylim(-300,1200)
plt.show()

fig, ax = plt.subplots()

ax.bar(labels_exp13, tot_slr_exp13, label=labels_exp13, align='edge', width=0.5)

ax.set_ylabel('mm SLR')
ax.set_title('ISMIP6 AIS EXP13 (PIGL Melt) 2100 SLR')

plt.xticks(rotation = 45, rotation_mode='anchor', ha='right')
plt.ylim(-300,1200)
plt.show()
```

```python
# Make dataframes of fixed/moving termini and PIGL/std melt.

fixed_fronts = df_pars.loc[df_pars["Ice Front"] == "Fix", "Participant"]
moving_fronts = df_pars.loc[df_pars["Ice Front"] != "Fix", "Participant"]

fixed_stdMelt = df_res.loc[df_res["Center+Model"].isin(fixed_fronts) & (df_res["Exp"] == "exp05")]
moving_stdMelt = df_res.loc[df_res["Center+Model"].isin(moving_fronts) & (df_res["Exp"] == "exp05")]
fixed_piglMelt =  df_res.loc[df_res["Center+Model"].isin(fixed_fronts) & (df_res["Exp"] == "exp13")]
moving_piglMelt =  df_res.loc[df_res["Center+Model"].isin(moving_fronts) & (df_res["Exp"] == "exp13")]

print(fixed_stdMelt)
print()
print(moving_stdMelt)
print()
print(fixed_piglMelt)
print()
print(moving_piglMelt)
```

```python
# Get total SLR for each model output

totals = pd.DataFrame()
totals["fixed_stdMelt"] = (list(fixed_stdMelt["Tot"]))
totals = pd.concat([totals,pd.Series(list(moving_stdMelt["Tot"]))], ignore_index=True, axis=1)
totals = pd.concat([totals,pd.Series(list(fixed_piglMelt["Tot"]))], ignore_index=True, axis=1)
totals = pd.concat([totals,pd.Series(list(moving_piglMelt["Tot"]))], ignore_index=True, axis=1)

totals.rename(columns = {0:"fixed_stdMelt"}, inplace = True)
totals.rename(columns = {1:"moving_stdMelt"}, inplace = True)
totals.rename(columns = {2:"fixed_piglMelt"}, inplace = True)
totals.rename(columns = {3:"moving_piglMelt"}, inplace = True)

# go from ice volume to water volume, divide by surface area of ocean, convert from m to mm
totals = -totals * (917/1000) * (1/(.71*np.pi*(6370e3)**2)) * 1000

print(totals)
```

```python
# Plot total SLR by category

totals_flip = totals
for col in totals_flip:
    totals_flip[col] = totals_flip[col].sort_values(ignore_index=True)

totals_flip = totals_flip.transpose()
print(totals_flip)

ax = totals_flip.plot.bar(rot=0)
ax.set_ylabel("SLR (mm)")
```

```python
# Calculate total and per category statistics

means = np.mean(totals)
print("Category Means (mm SLR)")
print(means)

SS_wc = np.sum((totals - np.mean(totals))**2)
print("\nPer category sum of squared deviation:")
print(SS_wc)

grand_mean = np.mean(np.mean(totals))
n_tot = totals.size
print("\nGrand mean: %.1f mm" % grand_mean)
print("Sample size all categories: %d" % n_tot)

SS_total = np.sum(np.sum(((totals - grand_mean)**2)))
print("\nTotal sum of squared deviation (treat all as one category): %.4s" % SS_total)

n_cat = totals.count()
print("\n Sample size per category:")
print(n_cat)

s_x = np.sqrt(SS_wc/(n_cat-1))
print("\nStandard error:")
print(s_x)

s_xBar = s_x/np.sqrt(n_cat)
print("\nStandard error of means:")
s_xBar
```

```python
# Plot Category mean and variance (error) with melt forcing on x axis

import matplotlib.pyplot as plt

fig, (ax0) = plt.subplots(nrows=1)
ax0.errorbar(["standard melt", "pigl melt"],
             np.array([means["fixed_stdMelt"],
                       means["fixed_piglMelt"]]),
             yerr=np.array([s_xBar["fixed_stdMelt"],s_xBar["fixed_piglMelt"]]), fmt='-o')
ax0.errorbar(["standard melt", "pigl melt"],
             np.array([means["moving_stdMelt"],
                       means["moving_piglMelt"]]),
             yerr=np.array([s_xBar["moving_stdMelt"],s_xBar["moving_piglMelt"]]), fmt='-o')
ax0.legend(["fixed termini", "calving law terminus movement"], loc='upper left')
ax0.set_ylabel("SLR (mm)")
```

```python
# Plot Category mean and variance (error) with calving behavior on x axis

fig, (ax1) = plt.subplots(nrows=1)
ax1.errorbar(["fixed terminus", "calving law"],
             np.array([means["fixed_stdMelt"],
                       means["moving_stdMelt"]]),
             yerr=np.array([s_xBar["fixed_stdMelt"],s_xBar["moving_stdMelt"]]), fmt='-o')
ax1.errorbar(["fixed terminus", "calving law"],
             np.array([means["fixed_piglMelt"],
                       means["moving_piglMelt"]]),
             yerr=np.array([s_xBar["fixed_piglMelt"],s_xBar["moving_piglMelt"]]), fmt='-o')
ax1.legend(["standard melt", "PIGL melt"], loc='upper left')
ax1.set_ylabel("SLR (mm)")
```

```python
SS_wc_tot = np.sum(SS_wc)
print("Sum of per category sum of squared deviance: %.0f" % SS_wc_tot)

df_wc_tot = totals.size - len(totals.columns)
MS_wc = SS_wc_tot/df_wc_tot
print("Variance within categories: %.0f" % MS_wc)
```

```python
# Determine significance of fixed / moving termini

import scipy.stats

fixed_mean = np.nanmean(
    np.concatenate((np.array(totals.fixed_stdMelt), np.array(totals.fixed_piglMelt))))
print("Fixed terminus mean %.1f" % fixed_mean)

moving_mean = np.nanmean(
    np.concatenate((np.array(totals.moving_stdMelt), np.array(totals.moving_piglMelt))))
print("Calving law mean %.1f" % moving_mean)

SS_calving = (n_cat.fixed_stdMelt+n_cat.fixed_piglMelt)*(fixed_mean-grand_mean)**2 + (
    n_cat.moving_stdMelt+n_cat.moving_piglMelt)*(moving_mean-grand_mean)**2
print("Sum of squared deviance for categories being fixed and moving terminus: %.0f" % SS_calving)

df_calving = 2-1 # deg of freedom is num groups - 1. We have two rows.
MS_calving = SS_calving / df_calving
print("Variance of rows (calving): %.0f" % MS_calving)

F_stat_calving = MS_calving/MS_wc
print("F statistic for calving yes/no: %.3f" % F_stat_calving)

F_sig_calving = scipy.stats.f.ppf(q=.95, dfn=df_calving, dfd=df_wc_tot)
print("F statistic for significance at 95%% confidence: %.3f" % F_sig_calving)

p_calving = 1-scipy.stats.f.cdf(F_stat_calving, df_calving, df_wc_tot)
print("Calving significance confidence level (p): %.3f" % p_calving)
```

```python
# Determine significance of melt forcing

stdMelt_mean = np.nanmean(
    np.concatenate((np.array(totals.fixed_stdMelt), np.array(totals.moving_stdMelt))))
print("Standard melt mean %.1f" % stdMelt_mean)

piglMelt_mean = np.nanmean(
    np.concatenate((np.array(totals.fixed_piglMelt), np.array(totals.moving_piglMelt))))
print("PIGL melt mean %.1f" % piglMelt_mean)

SS_melt = (n_cat.fixed_stdMelt+n_cat.moving_stdMelt)*(stdMelt_mean-grand_mean)**2 + (
    n_cat.fixed_piglMelt+n_cat.moving_piglMelt)*(piglMelt_mean-grand_mean)**2
print("Sum of squared deviance for categories being std and pigl melt: %.0f" % SS_melt)

df_melt = 2-1 # deg of freedom is num groups - 1. We have two rows.
MS_melt = SS_melt / df_melt
print("Variance of columns (melt forcing): %.0f" % MS_melt)

F_stat_melt = MS_melt/MS_wc
print("F statistic for melt std/PIGL: %.3f" % F_stat_melt)

F_sig_melt = scipy.stats.f.ppf(q=.95, dfn=df_melt, dfd=df_wc_tot)
print("F statistic for significance at 95%% confidence: %.3f" % F_sig_melt)

p_melt = 1-scipy.stats.f.cdf(F_stat_melt, df_melt, df_wc_tot)
print("Melt forcing significance confidence level (p): %.3f" % p_melt)
```

```python
# Test significance of coupled effect

SS_MxC = SS_total-SS_wc_tot-SS_melt-SS_calving
print("Sum of squared deviance for interaction: %.0f" % SS_MxC)

df_MxC= (2-1)*(2-1) # num rows - 1 times num cols - 1
MS_MxC = SS_MxC/df_MxC
print("Variance for inteaction: %.0f" % MS_MxC)

F_stat_MxC = MS_MxC/MS_wc
print("F statistic for interaction: %.3f" % F_stat_MxC)

F_sig_MxC = scipy.stats.f.ppf(q=.95, dfn=df_MxC, dfd=df_wc_tot)
print("F statistic for signficance at 95%% confidence: %.3f" % F_sig_MxC)

p_MxC = 1-scipy.stats.f.cdf(F_stat_MxC, df_MxC, df_wc_tot)
print("Interaction significance confidence level (p): %.3f" % p_MxC)
```
